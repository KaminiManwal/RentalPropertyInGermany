## DATA CONSISTENCY CHECKS

### a. Checking for mixed type data

```python
# Checking for mixed type data
for col in df_drop.columns.tolist():
  weird = (df_drop[[col]].applymap(type) != df_drop[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_drop[weird]) > 0:
    print (col)
```
