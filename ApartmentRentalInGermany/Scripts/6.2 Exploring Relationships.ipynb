## INVESTIGATING OUTLIERS

```python
sns.boxplot(df['totalRent'])
```

```python
# removing outlier from 'totalRent' column
upper_limit=25000
lower_limit=df['totalRent'].mean()-3*df['totalRent'].std()
print('upper limit:',upper_limit)
print('lower limit:',lower_limit)
```

```python
df.loc[(df['totalRent']>upper_limit) | (df['totalRent']<lower_limit)]
```

```python
df1=df.loc[(df['totalRent']<upper_limit) & (df['totalRent']>lower_limit)]
print('Data before removing outlier:',len(df))
print('Data after removing outlier:',len(df1))
```

```python
sns.boxplot(df1['totalRent'])
```

```python
# removing outlier from 'livingSpace' column
sns.boxplot(df1['livingSpace'])
```

```python
# removing outlier from livingSpace column
upper_limit1=1000
lower_limit1=df1['livingSpace'].mean()-3*df1['livingSpace'].std()
print('upper limit:',upper_limit1)
print('lower limit:',lower_limit1)
```

```python
df1.loc[(df1['livingSpace']>upper_limit1) | (df1['livingSpace']<lower_limit1)]
```

```python
df2=df1.loc[(df1['livingSpace']<upper_limit1) & (df1['livingSpace']>lower_limit1)]
print('Data before removing outlier:',len(df1))
print('Data after removing outlier:',len(df2))
```

```python
sns.boxplot(df2['livingSpace'])
```

```python
sns.boxplot(df2['yearConstructed'])
```

```python
# removing outlier from 'yearConstructed' column
upper_limit2=2022
lower_limit2=1800
print('upper limit:',upper_limit2)
print('lower limit:',lower_limit2)
```

```python
df2.loc[(df2['yearConstructed']>upper_limit2) | (df2['yearConstructed']<lower_limit2)]
```

```python
df3=df2.loc[(df2['yearConstructed']<upper_limit2) & (df2['yearConstructed']>lower_limit2)]
print('Data before removing outlier:',len(df2))
print('Data after removing outlier:',len(df3))
```

```python
df3.corr()
```

```python
df_sub1=df3[['totalRent','livingSpace','numberOfRooms','yearConstructed','lift','balcony','newlyConstructed']]
```

```python
# recreating correlation heatmap after removing outliers.
f,ax = plt.subplots(figsize=(10,10))
corr = sns.heatmap(df_sub1.corr(), annot = True, ax = ax)
```

I have re-run the correlation heatmap to observe the impact of outlier removal. The results indicate that 'total rent' is influenced by factors such as living space, year of construction, and additional amenities like a lift and balcony. Notably, there exists a strong correlation between 'totalRent' and 'livingSpace,' signifying that an increase in living space is associated with a corresponding increase in total rent.

```python
# exporting viz
corr.figure.savefig(os.path.join(path, 'Visualization', 'correlation.png'))
```

```python
# create scatter plot between year of construction and total rent
scatter_year=sns.lmplot(x = 'yearConstructed', y = 'totalRent', data = df3)
```

The scatter plot above illustrates the relationship between 'totalRent' and 'yearConstructed'. The trend line isn't visible due to the clustering of data points. As the years increase, more data clusters become apparent, suggesting a potential positive linear trend.

```python
# create scatterplot betwwen 'livingSpace' and 'totalRent'
scatter_living=sns.lmplot(x = 'livingSpace', y = 'totalRent', data = df3)
```

The displayed scatter plot illustrates the relationship between 'totalRent' and 'livingSpace', revealing a positive linear correlation between these two variables. ## 6. Categorical plot

```python
# Categorical plot
sns.histplot(df3['totalRent'], bins =20, kde = True)
```

```python
df3.loc[df3['livingSpace'] < 100, 'Rent_livingSpace'] = 'Low price'
```

```python
df3.loc[(df3['livingSpace'] >= 100) & (df3['livingSpace'] < 300), 'Rent_livingSpace'] = 'Middle price'
```

```python
df3.loc[df3['livingSpace'] >= 300, 'Rent_livingSpace'] = 'High price'
```

```python
df3['Rent_livingSpace'].value_counts(dropna = False)
```

```python
# Create a categorical plot in seaborn using the living space categories created above

sns.set(style="ticks")
cat_1= sns.catplot(x='state', y='livingSpace', hue='Rent_livingSpace', data=df3)
```

In the above categorical plot, it's clear that houses with substantial living spaces (over 300) and high prices are rare across all states represented on the x-axis. Conversely, there is a notable abundance of houses with living spaces under 100 square units, which appears consistently across all states. This range corresponds to the highest number of houses. Meanwhile, the mid-price range is low in comparison to the frequency of houses falling within the low price range.

```python
 # exporting viz
cat_1.figure.savefig(os.path.join(path, 'Visualization', 'categorical_livingSpace.png'))
```

```python
df3.loc[df3['yearConstructed'] < 1900, 'Rent_yearConstructed'] = 'Low price1'
```

```python
# Create a categorical plot in seaborn using the living space categories created above

sns.set(style="ticks")
cat_2 = sns.catplot(x='state', y='yearConstructed', hue='Rent_yearConstructed', data=df3)
```

Based on the categorical plot, it's evident that a majority of houses falling within the low price range are those constructed before the year 1900. The highest concentration of houses lies within the middle price range, aligning with properties built between 1900 and 2000. Comparatively, houses in the high price range are fewer in number, representing a smaller portion of the dataset.

```python
 # exporting viz
cat_2.figure.savefig(os.path.join(path, 'Visualization', 'categorical_yearConstructed.png'))
```
